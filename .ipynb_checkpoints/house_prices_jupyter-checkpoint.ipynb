# Initial Outlier Detection

From intro stats lets check the data with the 68-95-99.7 rule with respect to one-, two-, and three-standard deviations

```python
def trim(df,std_dev=1, scale = False, return_df = False):
    mu = df.SalePrice.mean()
    sigma = df.SalePrice.std()
    if scale:
        scale = 100000
    else:
        scale = 1
    trimmed_df = df.loc[(abs(df.SalePrice) <= (mu + std_dev*sigma))]
    data = trimmed_df.SalePrice
    if return_df:
        return trimmed_df
    else:
        label = str(std_dev) + "-Sigma, " + str(round(100*data.shape[0]/df.SalePrice.shape[0],2)) + "% of dataset"
        sbn.distplot(data/scale, kde_kws = {"label": label})
    

sbn.distplot(df.SalePrice/100000, kde_kws={"label": "All data"})
trim(df,1,True)
trim(df,2,True)
trim(df,3,True)
```

```python
trim(df,1,True)
trim(df,2,True)
trim(df,3,True)
plt.title('Scaled Standard Deviations')
```

To retain most of our observations, we subset the data on records within 2 standard deviations of the mean.  Next we'll try to build our model with strictly numerical features and observe how we do.

```python
df = trim(df,2,return_df = True)
```

# Numerical Features only
## Feature Selection

First we try to reduce the number of numerical features we have by testing for correlation.

```python
def abs_corr(df, drop_cols = [], min_val = .6, max_val = 1, plot_me = True, plot_x = 12, plot_y = 9):
    if len(drop_cols) > 0:
        drop_cols = list(set(drop_cols))
        df = df[[i for i in df.columns if i not in drop_cols]]
    abs_mtx = df.select_dtypes(exclude = ['object']).drop(columns = 'SalePrice').corr().abs()
    
    mtx = abs_mtx[(abs_mtx < max_val) & (abs_mtx >= min_val)].dropna(1, 'all').dropna(0, 'all')
    if plot_me:
        plt.subplots(figsize=(plot_x, plot_y))
        sbn.heatmap(mtx, vmax=.8, square=True)
    else:
        return mtx
```

Specifically, we white-out any correlations that fall below our threshold of .6 and observe the result below:

```python
abs_corr(df, plot_x = 10, plot_y = 7)
```

```python
# Lets examine further the relationship between GarageYrBlt and [YearBuilt, YearRemodAdd]
sbn.scatterplot('YearBuilt', 'GarageYrBlt',data = df)
print(df.loc[df['YearBuilt'] > df['GarageYrBlt'],['YearBuilt','GarageYrBlt']])
```

## Feature Engineering

Now that we have our numerical features of interest, let's see how many missings we have, and whether we can create new features:

Even though some features may be categorized as "numerical" by pandas, they may still be categorical in that they are not continuous (think a rating system 1-10). Excluding yearly variables, let's further examine:

```python
for col in df.select_dtypes(exclude = ['object']).drop(columns = 'SalePrice'):
    if len(df[col].unique()) < 12:
        print('Feature: ' + col)
        print(df[col].value_counts())
        print('\n')
```

Out of laziness, lets just turn them into dummy variables minus YrSold and PoolArea and go straight into modelling

```python
def reg_prep(df):
    for col in df.select_dtypes(exclude = ['object']):
        if col not in ['YrSold', 'PoolArea'] and len(df[col].unique()) < 12:
            print('The col to be dummied is ' + col)
            #print(pd.get_dummies(df[col],prefix = col))
            df = pd.concat([df.drop(columns = col),pd.get_dummies(df[col],prefix = col)], axis = 1)
        return df.select_dtypes(exclude = ['object'])


y = pd.concat([df['SalePrice'].to_frame(), pd.read_csv('data/sample_submission.csv').set_index('Id')])

test = pd.read_csv('data/test.csv').set_index('Id')
df.drop(columns = 'SalePrice', inplace = True)
X = reg_prep(pd.concat([df,test[df.columns]]))
```

```python
def check_skew_kurtosis(df, feature = 'SalePrice', pics_only = False,):
    y = df[feature]
    sbn.distplot(y, fit=norm)
    plt.figure()
    stats.probplot(y, plot=plt)
    plt.show()
    print('The kurtosis: ' + str(stats.kurtosis(y)))
    print('The skew: ' + str(stats.skew(y)))
```

```python
continuous = ['LotArea', 'BsmtFinSF1', 'BsmtFinSF2', 'BsmtUnfSF', '1stFlrSF','LowQualFinSF','GrLivArea','GarageArea',
              'WoodDeckSF', 'OpenPorchSF', 'EnclosedPorch','3SsnPorch','ScreenPorch', 'PoolArea']

for col in continuous:
    X[col] = (X[col] - X[col].mean())/X[col].std()
```
